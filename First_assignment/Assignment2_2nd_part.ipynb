```python
import numpy as np
```

```python
mean=[0, 10, 5, 20, 15]
variance=[1,2,4,1.2,7]
```

# Collecting Training and Test samples 

```python
sam=np.random.choice(5, 50000, p=[0.1,0.25,0.3,0.25,0.1])
test_sam=np.random.choice(5, 10000, p=[0.1,0.25,0.3,0.25,0.1])
```

```python
estimated_priors=[0,0,0,0,0]
estimated_test_priors=[0,0,0,0,0]
training_samples=[]
samples_in_class=[0,0,0,0,0]
test_samples=[]
estimated_mean=[int(0),int(0),int(0),int(0),int(0)]
estimated_variances=[]
for i in range(50000):
    training_samples.append([np.random.normal(mean[sam[i]], np.sqrt(variance[sam[i]]),1),sam[i]])
for i in range(10000):
    test_samples.append([np.random.normal(mean[test_sam[i]], np.sqrt(variance[test_sam[i]]),1),test_sam[i]])
    estimated_test_priors[test_samples[i][1]]+=1
for i in range(50000):
    estimated_mean[sam[i]]=training_samples[i][0]+estimated_mean[sam[i]]
    estimated_priors[sam[i]]=1+estimated_priors[sam[i]]
for i in range(5):
    estimated_mean[i]=estimated_mean[i]/estimated_priors[i]
```
