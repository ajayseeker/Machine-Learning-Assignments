# Creating a list of transactions and individual items

```python
transactions=[]
l=[]
items=[]
with open('store_data.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for row in csv_reader:
        for i in row:
            if(i!=""):
                l.append(i)
                items.append(i)
        transactions.append(l)
        l=[]
```

##  Writing a function to calculate support for a list containing items

```python
def Support(l, trans):
    count=0
    flag= True
    for t in trans:
        for i1 in l:
            if(i1 in t):
                continue
            else:
                flag=False
                break
        if(flag==True):
            count+=1
        else:
            flag=True
    return count/len(trans)
```

# Making a list of all the 1 element which has minimum support 

```python
set1=[]
for i in items:
    if(Support([i], transactions)>min_support):
        set1.append(i)
```

#  Saving all two elements sets with minimum support

```python
set_temp=list(itertools.combinations(set1,2))
temp=[]
t=[]
set2=[]

for i in range(len(set_temp)):
    for k in set_temp[i]:
        temp.append(k)
    if(Support(temp, transactions)<min_support):
        t.append(set_temp[i])
    temp=[]
for i in t:
    set_temp.remove(i)
t=[]
for i in range(len(set_temp)):
    for j in set_temp[i]:
        t.append(j)
    set2.append(t)
    t=[]
print(set2)
```

# Using two element sets with minimum support and one element sets with minimum support to make list of all bigger sets of items which has minimum support 

```python
freq=[]
f=[0]
set3=[]
temp=[]
r=[]
freq.append(set2)
count=0
p=0
while(len(f)!=0):
    f=[]
    set3=freq[count]
    for i in range(len(set3)):
        for j in set1:
            if((j in set3[i])==False):
                t=set3[i].copy()
                t.append(j)
            else:
                continue
            if(len(t)==3):
                p+=1
            if(Support(t, transactions)>min_support):
                f.append(t)
                
    if(len(f)>0):
        freq.append(f)
    count+=1
print(p) 
```

# Function to find confidence 

```python
def Confidence(l, trans):
    count=0
    s=0
    flag = True
    for k in trans:
        for i in l[0]:
            if((i in k)==False):
                flag=False
                break
        if(flag==True):
            s+=1
            if(l[1][0] in k):
                count+=1
        else:
            flag=True
    return count/s        
```

```python
min_confidence=0.05
rules=[]
temp=[[],[]]
for i in range(len(freq)):
    for j in freq[i]:
        for l in range(len(j)):
            for k in range(len(j)):
                if(k==l):
                    temp[1].append(j[l])
                else:
                    temp[0].append(j[k])
            if((len(temp[0])==0) or (len(temp[1])==0)):
                    continue
            if(Confidence(temp, transactions)>min_confidence):
                    rules.append(temp)
            
            temp=[[],[]]
```

# Final rules
